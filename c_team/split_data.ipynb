7천 글자 이상 법령을 따로 분류
'장'으로 나누어져 있는 법령, '조'로만 나누어져 있는 법령 두가지로 분류함.

```python
from pathlib import Path

p = Path('./data')
long_files = []
long_file_group=[]
for file_path in p.iterdir():
    with open(file_path, 'r', encoding='utf-8') as f:
        text = f.read()
        if len(text) > 7000:
            if '제1장' in text:
                long_file_group.append(file_path)
            else:
                long_files.append(file_path)
```

```python
len(long_file_group), len(long_files)
```

'장'으로 구분되어있는 법령은 '장'으로 잘라서 분할 저장

```python
import re
from tqdm import tqdm

for file_path in tqdm(long_file_group, 'Splitting long files'):
    file_name = str(file_path).split('/')[-1].replace('.txt', '')
    pattern = r'(?=제\s*\d+\s*장)'

    with open(file_path, 'r', encoding='utf-8') as f:
        text = f.read()

    result = re.split(pattern, text)
    result = [c.strip() for c in result if c.strip()]

    for idx, chunk in enumerate(result):
        with open(f'data/{file_name}_{idx+1}.txt', 'w', encoding='utf-8') as f:
            f.write(chunk)
```

'조'로만 구성된 문서는 2개로 나눠서 저장

```python
pattern =r'(?=제\s*\d+\s*조(?:의\s*\d+)?\s*\([^)]+\))'

for file_path in tqdm(long_files, 'long files split'):
    file_name = str(file_path).split('/')[-1].replace('.txt', '')

    with open(file_path, 'r') as f:
        text = f.read()
    articles = re.findall(pattern, text)
    size = len(articles) // 2

    result = re.split(pattern, text)
    result = [c.strip() for c in result if c.strip()]

    start = ' '.join(result[:size])
    end = ' '.join(result[size:])
    chunks = [start, end]

    for idx, chunk in enumerate(chunks):
        with open(f'data/{file_name}_{idx+1}.txt', 'w', encoding='utf-8') as f:
          f.write(chunk)
```
